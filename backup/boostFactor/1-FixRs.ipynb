This model is not only a fit of the Boost-Factor. You also need to consider the fit of the two linear relations assumed, $b_0(\lambda,z), R_s(\lambda, z)$, with an intrisic scatter and correlation. My quick and dirty approach was to take an additional scatter of the global relation. The results as shown above works well for the diagonal terms of the $\lambda,z$ bins, i.e. low(high)z or low(high) lambda. The off diagonal bins don't work because there's a correlation of the boost factor amplitude with the lambda and redshift. Here we assumed both are uncorrelated linearly relations.

````python
def Boost_Factor_param(Z,L,Rs,alpha_Rs,beta_Rs,B0,alpha_B0,beta_B0):
    """This function takes the priors from the values file and calculates rs and b0 values for the model given redshift and lambda
    """
    rs =Rs*(Z/(1+0.3))**alpha_Rs*(L/30.)**beta_Rs
    b0= B0*((Z)/(1+0.3))**alpha_B0*(L/30.)**beta_B0
    return rs,b0

````

````python
def _Boost_Factor_fit_rs(x, Rs0, alpha_Rs, beta_Rs):
    """This function takes the priors from the values file and calculates rs and b0 values for the model given redshift and lambda
    """
    # unpack variables
    Z, L = x
    rslog = np.log(Rs0) + alpha_Rs*np.log(Z/(1+0.3)) + beta_Rs*np.log(L/30.)
    return rslog

def _Boost_Factor_fit_b0(x, B0, alpha_B0, beta_B0):
    """This function takes the priors from the values file and calculates rs and b0 values for the model given redshift and lambda
    """
    # unpack variables
    Z, L = x
    b0log = np.log(B0) + alpha_B0*np.log(Z/(1+0.3)) + beta_B0*np.log(L/30.)
    return b0log
````
